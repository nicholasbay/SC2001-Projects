```python
def knapsack(C, n, weights, profits):
#     Computes the maximum profit that can be made by packing objects into a knapsack of capacity C, given n types of objects with weights wi and profits pi.

#     Args:
#         C: The capacity of the knapsack.
#         weights: A list of the weights of the objects.
#         profits: A list of the profits of the objects.

#     Returns:
#         The maximum profit that can be made.

    table=[]
    
    for i in range(C+1):
        table.append(0)
        
    for i in range(0, C+1):
        print("\n\nAt capacity [",i,"]:")
        for j in range(0, n):
            if weights[j] <= i:
                table[i] = max(table[i] , table[i - weights[j]] + profits[j])
                print("After considering item[", j, "], weight: ", table[i])
    return table[C]
```

# RESULTS FOR PART 4(a)
